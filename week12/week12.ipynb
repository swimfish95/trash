```cpp
#include <stdio.h>

int MYPOWER(int a,int b)
{   int ans=1;
    for (int i=0;i<b;i++){
        ans =ans*a;
    }
    return ans;
}

int main(void)
{
	int a,b;
	scanf("%d%d",&a,&b);
	printf("[%d]",MYPOWER(a,b));
	return 0;
}
```

```cpp
// week12-4.cpp
#include <stdio.h>

int main()
{
    int a, b;
    scanf("%d%d", &a, &b);
    
    int ans = 1;
    

    for(int i=1; i<=a; i++){
        // 如果 i 可以同時整除 a 和 b (也就是餘數都是 0)
        if(a % i == 0 && b % i == 0) {
            ans = i;
    }
    
    // 印出結果與約分
    printf("最大公因數是%d\n", ans);
    printf("%d/%d 約分之後,變%d/%d\n", a, b, a/ans, b/ans);

    return 0;
}
```
