```java
// week14-1 真正的 3x3 打地鼠 (支援滑鼠點擊)
PImage img;

void setup() {
  size(300, 300);
  img = loadImage("gopher.png");
}

int[][] a = {
  {0, 0, 0},
  {0, 0, 0},
  {0, 0, 0}
};

void draw() {
  background(#FFFFF2);
  fill(#FFFFF2); // 設定方形填色，讓被打死畫出來的方形跟背景顏色一樣
  
  // 1. 隨機冒出地鼠 (給予 60 幀的生命值)
  if (frameCount % 60 == 0) {
    int i = int(random(3)), j = int(random(3));
    a[i][j] = 60;
  }
  
  // 2. 繪製九宮格狀態
  for (int i=0; i<3; i++) {
    for (int j=0; j<3; j++) {
      float x = j * 100;
      float y = i * 100 + 100; // 預設 Y 座標在洞口下方
      
      if (a[i][j] > 0) {
        // 【狀態 A：活著的地鼠】
        a[i][j] -= 1; // 生命倒數
        y -= 100 * sin(a[i][j] * PI / 60); // 探頭動畫
        image(img, x, y, 100, 100);
        
      } else if (a[i][j] < 0) {
        // 【狀態 B：死掉的地鼠 (顯示大叉叉)】
        y -= 100; // 將 Y 座標推回對齊格子的左上角
        rect(x, y, 100, 100); // 畫一個方塊蓋住原本的位置
        
        // 畫一個大叉叉 X
        line(x, y, x + 100, y + 100);
        line(x + 100, y, x, y + 100);
        
        a[i][j] += 1; // 死亡倒數計時：從 -120 慢慢加回 0，到了 0 叉叉就會消失
      }
    }
    // 遮罩魔法：蓋住下半身，讓地鼠有從洞裡鑽出來的感覺
    rect(0, 100 + i*100, 300, 100);
  }
}

// 3. 滑鼠點擊事件 (打地鼠的關鍵！)
void mousePressed() {
  // 將滑鼠座標 (0~299) 除以 100，瞬間換算成陣列的索引值 (0, 1, 2)
  int i = mouseY / 100;
  int j = mouseX / 100;
  
  // 如果滑鼠點擊的這個格子，剛好有「活著的地鼠」
  if (a[i][j] > 0) {
    a[i][j] = -120; // 宣判死刑！賦予它負數的值，變成顯示 2 秒鐘 (120幀) 的大叉叉
  }
}
```

```cpp
//week14-3a
#include <stdio.h>

int main()
{
    char s[20]; // 宣告一個可以裝 20 個字元的字串陣列
    scanf("%s", s); // 讀取使用者輸入的字串 (遇到空白會停止)
    
   
    for (int i=0; s[i] != 0; i++) {
        
        
        if (s[i] >= 'A' && s[i] <= 'Z') {
            // 轉換成小寫
            s[i] = s[i] - 'A' + 'a';
        }
       
        else if (s[i] >= 'a' && s[i] <= 'z') {
            // 轉換成大寫
            s[i] = s[i] - 'a' + 'A';
        }
    }
    

    printf("%s\n", s);
    
    return 0;
}
```

```cpp
//week14-5a
#include <stdio.h>
int main() {
    int N;
    scanf("%d", &N);
    
    if (N < 0) N = -N;
    
    int ans = 0;
    while (N > 0) {
        ans = N % 10;
        N = N / 10;   
    }
   
    printf("%d\n", ans);
    
    return 0;
}
```

```cpp
// week14-6.cpp SOIT108_Advance_014b
#include <stdio.h>

int main()
{
    int N;
    scanf("%d", &N);
    
    
    int zeros = 1;
    
    while (N > 0) {
        // N % 10 抓出最後一個數字，再乘上它專屬的位值 (zeros)
        printf("%d ", (N % 10) * zeros);
        
        N = N / 10;         // 砍掉最後一位數
        zeros = zeros * 10;
    }
    
    return 0;
}
```
