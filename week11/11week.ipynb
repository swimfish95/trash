```cpp
// week11-2c.cpp SOIT106_ADVANCE_001
#include <stdio.h>

int main()
{
    int N;
    scanf("%d", &N);
    
    int ans = 0; // Part B：準備用來裝反轉數字的變數
    int N2 = N;  // Part C：🌟 關鍵備份！把原本的 N 存進 N2 保護起來
    
    // 開始剝皮與重組
    while (N>0) {
        ans = ans * 10 + N%10; // Part B：把剝下來的皮加進 ans
        N = N / 10;            // 這裡會把 N 一步步砍到變成 0
    }
    
    // 因為 N 已經陣亡變成 0 了，所以我們呼叫備份的 N2 來印出結果
    printf("%d+%d=%d\n", N2, ans, ans+N2);
    
    return 0;
}
```

```cpp
// week11-4.cpp SOIT108_Advance_001
#include <stdio.h>

int main()
{
    int N;
    scanf("%d", &N); // 讀取使用者輸入的數字 N
    
    int ans = 0; // 預設答案為 0
    
    // 從 1 開始，一路往上猜到 1000
    for (int i=1; i<=1000; i++) {
        // 如果猜的數字 i 乘上自己 (i*i) 剛好等於 N，那 i 就是 N 的平方根！
        if ( i*i == N ) {
            ans = i; // 把找到的正確答案存起來
        }
    }
    
    // 印出最終找到的答案
    printf("%d", ans);
    
    return 0; // 補上好習慣
}
```

```cpp
// week11-5.cpp SOIT108_Advance_002
#include <stdio.h>

int main()
{
    int a, b;
    // 讀取兩個數字 a 和 b，代表我們要尋找的範圍 (例如從 10 找 到 20)
    scanf("%d%d", &a, &b);
    
    int ans = 0; // 用來記錄「總共找到了幾個質數」
    
    // 外層迴圈 i：從 a 一路檢查到 b，每個數字都拿來測試看看
    for (int i=a; i<=b; i++) {
        
        int bad = 0; // 🚩 設立一個「壞寶寶旗標」，0 代表預設它是質數 (好寶寶)
        
        // 內層迴圈 k：用來測試 i 是不是質數
        // 質數的定義是「只能被 1 和自己整除」。所以我們從 2 開始除，一路除到 i-1
        for (int k=2; k<i; k++) {
            // 藍色反白區塊：如果 i 竟然可以被 k 整除 (餘數為 0)
            if (i % k == 0) {
                bad = 1; // 抓到了！它有其他的因數，把它標記為壞寶寶 (bad = 1)
            }
        }
        
        // 如果經歷了上面的層層考驗，bad 依然是 0，代表它真的是質數！
        if (bad == 0) {
            ans++; // 質數數量 + 1
        }
    }
    
    // 印出最終找到了幾個質數
    printf("%d\n", ans);
    
    return 0;
}
```

```cpp
// week11-6a.cpp SOIT108_Advance_002B
#include <stdio.h>

int main()
{
    int a, b, c;
    scanf("%d%d%d", &a, &b, &c); // 讀取三個數字

    // 1. 確保 a 比 b 大
    if (a < b) { // 大小不對，就交換
        int t = a; a = b; b = t;
    }

    // 2. 確保 a 比 c 大
    if (a < c) { // 大小不對，就交換
        int t = a; a = c; c = t;
    }
    // 經過上面兩關，a 已經和 b、c 都比過了，所以現在 a 絕對是全場最大！

    // 3. 確保 b 比 c 大
    if (b < c) { // 大小不對，就交換
        int t = b; b = c; c = t;
    }
    // 經過這關，b 確定大於 c。
    // 到這裡，三個數字已經完美按照 a(大) > b(中) > c(小) 排好了！

    // 4. 把排好的數字組合成一個三位數，再加 1
    printf("%d", a*100 + b*10 + c + 1);

    return 0; // 養成好習慣
}
```

```cpp
// week11-6b.cpp SOIT108_Advance_004
#include <stdio.h>

int main()
{
    int N;
    scanf("%d", &N); // 讀取使用者輸入的數字 N

    int ans = 0; // 準備一個撲滿 (ans)，一開始裡面是 0 元

    // 迴圈從 i = 1 一路數到 N
    for (int i=1; i<=N; i++) {
        // 把 i 乘上 11 後，丟進撲滿裡累加
        ans += i * 11;
    }

    // 印出撲滿裡最後的總金額
    printf("%d", ans);

    return 0; // 養成好習慣
}
```
